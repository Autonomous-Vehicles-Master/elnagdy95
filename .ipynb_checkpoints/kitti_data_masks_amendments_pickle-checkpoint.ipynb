```python
import numpy as np
import cv2
import glob
import os

#images = [cv2.imread(file) for file in glob.glob("path/to/files/*.png")]
#im = cv2.imread("ft.png")
#path = "E:\THI MASTERS\Semestre1\Electives\Autonomous Vehicles by Machine Learning Algorithms\Project\New folder\keras-yolo3-master\kitti_mask\KITTI\*.*"
inputfolder = r'E:\THI MASTERS\Semestre1\Electives\Autonomous Vehicles by Machine Learning Algorithms\Project\New folder\keras-yolo3-master\kitti_mask\KITTI'
folderlen = len(inputfolder)
os.mkdir('resized1')

for file in glob.glob(inputfolder + "/*.png"):
    im = cv2.imread(file)
    #print(im.shape)
    red_channel = im[:,:,2]
    rows,cols=red_channel.shape

    for i in range(rows):
        for j in range(cols):
            if red_channel[i,j] == 4 or red_channel[i,j]==3:
                red_channel[i,j] = 255
            else:
                red_channel[i,j]=0
    red_channel=cv2.resize(red_channel, (160, 80))
    
    rows,cols=red_channel.shape
    for i in range(rows):
        for j in range(cols):
            if red_channel[i,j] >0:
                red_channel[i,j] = 255
            else:
                red_channel[i,j]=0
    cv2.imwrite("resized1" + file[folderlen:] ,red_channel)
#cv2.imwrite('ft1.png',im)
```

```python
import numpy as np
import cv2
import glob
import os

#images = [cv2.imread(file) for file in glob.glob("path/to/files/*.png")]
#im = cv2.imread("ft.png")
#path = "E:\THI MASTERS\Semestre1\Electives\Autonomous Vehicles by Machine Learning Algorithms\Project\New folder\keras-yolo3-master\kitti_mask\KITTI\*.*"
inputfolder = r'E:\THI MASTERS\Semestre1\Electives\Autonomous Vehicles by Machine Learning Algorithms\Project\New folder\keras-yolo3-master\kitti_mask\training_dataset'
folderlen = len(inputfolder)
os.mkdir('resized3')

for file in glob.glob(inputfolder + "/*.png"):
    im = cv2.imread(file)
    
    im=cv2.resize(im, (160, 80))
    #print(im.shape)
    #red_channel = im[:,:,2]
    #rows,cols=red_channel.shape

    #for i in range(rows):
        #for j in range(cols):
            #if red_channel[i,j] == 4 or red_channel[i,j]==3:
                #red_channel[i,j] = 255
            #else:
                #red_channel[i,j]=0
                
    img_y_cr_cb = cv2.cvtColor(im, cv2.COLOR_BGR2YCrCb)
    y, cr, cb = cv2.split(img_y_cr_cb)

    # Applying equalize Hist operation on Y channel.
    #y_eq = cv2.equalizeHist(y)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    y_eq = clahe.apply(y)

    img_y_cr_cb_eq = cv2.merge((y_eq, cr, cb))
    im = cv2.cvtColor(img_y_cr_cb_eq, cv2.COLOR_YCR_CB2BGR)

    cv2.imwrite("resized3" + file[folderlen:] ,im)
#cv2.imwrite('ft1.png',im)
```
